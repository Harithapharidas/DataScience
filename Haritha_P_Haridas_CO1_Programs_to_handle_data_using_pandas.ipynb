```python
# 5. Given is a dataframe showing name, occupation,salary of people. Find the average salary per occupation.
import pandas as pd

detail={
  'Name':["Adarsh","Bhagya","Seetha","Haritha","Vishnu","Jissmon"],
  'Occuption':["Manager","Businnes","Manager","Business","ASE","ASE"],
  'Salary':[25000,15000,27000,24000,16000,23000]
}
df=pd.DataFrame(detail)
print(df)
occ=df.groupby("Occuption")['Salary'].mean()
print("Average Salary per Occuption")
print(occ)
```
